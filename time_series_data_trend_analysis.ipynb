```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "date": pd.to_datetime(["2022-02-01", "2022-03-01", "2022-04-01", "2022-05-01", "2022-06-01", "2022-08-01", "2022-09-01", "2023-01-01", "2023-02-01", "2023-03-01","2023-04-01","2023-05-01"]),
    "value": [10, 15, 22, 28, 35, 42, 48, 55, 60, 68,75,82]
}
df = pd.DataFrame(data)
df.set_index("date", inplace=True)

plt.figure(figsize=(10, 6))
plt.plot(df.index, df["value"], marker='o')
plt.xlabel("Date")
plt.ylabel("Value")
plt.title("Time Series data with Trend")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

date_range = pd.date_range(start='2024-01-01', end='2024-12-29', freq='D')
data = pd.Series(np.random.randn(len(date_range)), index=date_range)


decomposition = seasonal_decompose(data, model='additive')


plt.figure(figsize=(12, 8))


plt.subplot(2, 1, 1)
data.plot()
plt.title('Time Series Data')
plt.xlabel('Date')
plt.ylabel('Value')


plt.subplot(2, 1, 2)
seasonality = decomposition.seasonal
seasonality.plot()
plt.title('Seasonality')
plt.xlabel('Date')
plt.ylabel('Seasonal Component')

plt.tight_layout()
plt.show()

```
